## Fisher's P-value

```python
import scipy.stats as stats

# Contingency table:
#        Upregulated   Not Upregulated
# Sim    a             b
# NonSim c             d

# Number of upregulated and not-upregulated genes in both groups
a = len(Similar_SigUpregul_pcg_set)
b = len(similar_pcg_set) - a
c = len(Nonsimilar_SigUpregul_pcg_set)
d = len(nonsimilar_pcg_set) - c

# Create the contingency table
contingency_table = [[a, b], [c, d]]

# Perform Fisher's exact test
oddsratio, p_value = stats.fisher_exact(contingency_table)

print(f"P-value: {p_value}")
```
