```python
import random
iterations = 10000

low_then_expect = 0
high_then_expect = 0

for i in range(0, iterations):
    num = len(similar_pcg_set)
    Random_pcg_list = random.sample(pcg_set, num) 
    # num = len(similar_pcg_set-(similar_pcg_set-RNAseq_pcg_set_L))
    # Random_pcg_list = random.sample(RNAseq_pcg_set_L, num)          # change set
    
    Random_SigUpregul_pcg_list = []
    Random_SigUpregul_pcg_list.extend([gene_id for gene_id in Random_pcg_list if gene_id in SigUpregul_pcg_set])          ##
    
    if len(Random_SigUpregul_pcg_list) < len(Similar_SigUpregul_pcg_set):
        low_then_expect += 1
    else:
        high_then_expect += 1

print("low_then_expect: ", low_then_expect)
print("high_then_expect: ", high_then_expect)
```

```python
p_value = high_then_expect / iterations

print("Ranked P value: ", p_value)
```
